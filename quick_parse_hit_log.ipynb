```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
filename_to_parse = "../check_exp.log"
```

```python
def get_values(filename_to_parse, key):
    with open(filename_to_parse, "r") as f:
        data = f.readlines()
    values = [json.loads(line)[key] for line in data[1:]]
    return values
```

```python
df = sorted(get_values("../check_exp.log", "StartTimeNs"))
intervals = [df[i + 1] - df[i] for i in range(len(df) - 1)]
intervals
plt.hist(intervals, bins=100)
plt.xlim(0, 8*10e5)
plt.show()
```

```python
df = sorted(get_values("../check_none.log", "StartTimeNs"))
intervals = [df[i + 1] - df[i] for i in range(len(df) - 1)]
intervals
plt.hist(intervals, bins=100)
# plt.xlim(0, 8*10e7)
plt.show()
```

```python
none = np.array(sorted(get_values("../check_none.log", "StartTimeNs"))) / 10**6
none_intervals = [none[i + 1] - none[i] for i in range(len(none) - 1)]

exp = np.array(sorted(get_values("../check_exp.log", "StartTimeNs"))) / 10**6
exp_intervals = [exp[i + 1] - exp[i] for i in range(len(exp) - 1)]
```
